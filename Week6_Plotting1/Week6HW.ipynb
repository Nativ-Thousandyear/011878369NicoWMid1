## Problem 3

In the cell below make a line plot of the closing prices of the GME stock data and line plot of the daily high on the same set of axes. Give the plot a title and label the axes.  

```python
import matplotlib.pyplot as plt
import pandas as pd

# Manually create the DataFrame (replace with your actual data)
data = [
    ["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"],
    ["2/4/2020", 4.03, 4.25, 3.97, 4.07, 4.07, 3563100],
    # ... (remaining data rows)
]

df = pd.DataFrame(data[1:], columns=data[0])

# Ensure the "Date" column is in datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")

# Create line plots for closing and high prices
plt.plot(df["Date"], df["Close"], label="Closing Price", color="blue")
plt.plot(df["Date"], df["High"], label="Daily High", color="red")

# Customize plot
plt.title("GME Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()  # Add a legend

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame containing tuition information
tuition_df = pd.DataFrame({
    "WSU": [27302, 27697, 28206, 27825, 26899, 27249, 27484, 27991, 28520],
    "UI": [19700, 19598, 19630, 20294, 20640, 21300, 21350, 21820, 21820],
    "year": [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
})

# Scatter plot of WSU vs. UI tuition
plt.scatter(tuition_df['WSU'], tuition_df['UI'], c=tuition_df['year'], cmap='coolwarm')
plt.colorbar(label='Year')
plt.title('WSU vs. UI Tuition')
plt.xlabel('WSU Tuition')
plt.ylabel('UI Tuition')
plt.show()
```
